```python
import json
import os
from collections import Counter
```

```python
file_path = "/root/autodl-tmp/data/cxrgraph/pred_rel/test_mimic.json"
```

```python
with open(file_path, "r") as f:
    docs = [json.loads(lines) for lines in f]
```

```python
eval_results = {
    "ner": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "ner_anat": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele if ele[2] == "Anatomy"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "ner_obsp": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele if ele[2] == "Observation-Present"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "ner_obsa": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele if ele[2] == "Observation-Absent"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "ner_obsu": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele if ele[2] == "Observation-Uncertain"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "ner_locatt": {
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["ner"] for ele in sent_ele if ele[2] == "Location-Attribute"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr": {  # predicted entity types also must be correct
        "num_gold_label": len([i for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele for i in ele[2:] if i != "NA"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_normality": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[2] != "NA"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_nor": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[2] == "Normal"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_abnor": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[2] == "Abormal"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_action": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[3] != "NA"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_rm": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[3] == "Removable"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_ess": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[3] == "Essential"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_change": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[4] != "NA"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_pos": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[4] == "Positive"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_neg": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[4] == "Negative"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "attr_unch": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["entity_attributes"] for ele in sent_ele if ele[4] == "Unchanged"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel+": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel+modify": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele if ele[4] == "modify"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel+locat": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele if ele[4] == "located_at"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel+sugg": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele if ele[4] == "suggestive_of"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
    "rel+part": {  # predicted entity types also must be correct
        "num_gold_label": len([ele for doc in docs for sent_ele in doc["relations"] for ele in sent_ele if ele[4] == "part_of"]),
        "num_pred_label": 0,
        "num_correct_label": 0,
    },
}

for doc in docs:
    #ent
    all_ners = [ner for sent in doc["ner"] for ner in sent]
    all_pred_ners = [ner for sent in doc["pred_ner"] for ner in sent]
    corr_pred_ner_indices = []
    for pred_ner in all_pred_ners:
        key = "num_pred_label"
        eval_results["ner"]["num_pred_label"] += 1
        if pred_ner[2] == "Anatomy":
            eval_results["ner_anat"][key] += 1
        if pred_ner[2] == "Observation-Present":
            eval_results["ner_obsp"][key] += 1
        if pred_ner[2] == "Observation-Absent":
            eval_results["ner_obsa"][key] += 1
        if pred_ner[2] == "Observation-Uncertain":
            eval_results["ner_obsu"][key] += 1
        if pred_ner[2] == "Location-Attribute":
            eval_results["ner_locatt"][key] += 1
        if pred_ner in all_ners:
            key = "num_correct_label"
            corr_pred_ner_indices.append((pred_ner[0],pred_ner[1]))
            eval_results["ner"]["num_correct_label"] += 1
            if pred_ner[2] == "Anatomy":
                eval_results["ner_anat"][key] += 1
            if pred_ner[2] == "Observation-Present":
                eval_results["ner_obsp"][key] += 1
            if pred_ner[2] == "Observation-Absent":
                eval_results["ner_obsa"][key] += 1
            if pred_ner[2] == "Observation-Uncertain":
                eval_results["ner_obsu"][key] += 1
            if pred_ner[2] == "Location-Attribute":
                eval_results["ner_locatt"][key] += 1

    # attr
    all_attrs = [attr for sent in doc["entity_attributes"] for attr in sent]
    all_attr_indices = [(attr[0],attr[1]) for sent in doc["entity_attributes"] for attr in sent]
    all_pred_attrs = [attr for sent in doc["pred_attributes"] for attr in sent]
    for pred_attr in all_pred_attrs:
        key = "num_pred_label"
        if pred_attr[2] != "NA":
            eval_results["attr"][key] += 1
            eval_results["attr_normality"][key] += 1
        if pred_attr[2] == "Normal":
            eval_results["attr_nor"][key] += 1
        if pred_attr[2] == "Abnormal":
            eval_results["attr_abnor"][key] += 1
        if pred_attr[3] != "NA":
            eval_results["attr"][key] += 1
            eval_results["attr_action"][key] += 1
        if pred_attr[3] == "Removable":
            eval_results["attr_rm"][key] += 1
        if pred_attr[3] == "Essential":
            eval_results["attr_ess"][key] += 1
        if pred_attr[4] != "NA":
            eval_results["attr"][key] += 1
            eval_results["attr_change"][key] += 1
        if pred_attr[4] == "Positive":
            eval_results["attr_pos"][key] += 1
        if pred_attr[4] == "Negative":
            eval_results["attr_neg"][key] += 1
        if pred_attr[4] == "Unchanged":
            eval_results["attr_unch"][key] += 1
        
        if (pred_attr[0],pred_attr[1]) in all_attr_indices:
            key = "num_correct_label"
            gold_attr = all_attrs[all_attr_indices.index((pred_attr[0],pred_attr[1]))]
            
            if pred_attr[2] != "NA" and pred_attr[2] == gold_attr[2]:
                eval_results["attr"][key] += 1
                eval_results["attr_normality"][key] += 1
                if pred_attr[2] == "Normal":
                    eval_results["attr_nor"][key] += 1
                if pred_attr[2] == "Abnormal":
                    eval_results["attr_abnor"][key] += 1
                    
            if pred_attr[3] != "NA" and pred_attr[3] == gold_attr[3]:
                eval_results["attr"][key] += 1
                eval_results["attr_action"][key] += 1
                if pred_attr[3] == "Removable":
                    eval_results["attr_rm"][key] += 1
                if pred_attr[3] == "Essential":
                    eval_results["attr_ess"][key] += 1
                
            if pred_attr[4] != "NA" and pred_attr[4] == gold_attr[4]:
                eval_results["attr"][key] += 1
                eval_results["attr_change"][key] += 1
                if pred_attr[4] == "Positive":
                    eval_results["attr_pos"][key] += 1
                if pred_attr[4] == "Negative":
                    eval_results["attr_neg"][key] += 1
                if pred_attr[4] == "Unchanged":
                    eval_results["attr_unch"][key] += 1
    
    # rel
    all_rels = [ner for sent in doc["relations"] for ner in sent]
    all_pred_rels = [ner for sent in doc["pred_rel"] for ner in sent]
    for pred_rel in all_pred_rels:
        key = "num_pred_label"
        eval_results["rel"][key] += 1
        eval_results["rel+"][key] += 1
        subj = (pred_rel[0], pred_rel[1])
        obj = (pred_rel[2], pred_rel[3])
        rel_type = pred_rel[4]
        if rel_type == "modify":
            eval_results["rel+modify"][key] += 1
        if rel_type == "located_at":
            eval_results["rel+locat"][key] += 1
        if rel_type == "suggestive_of":
            eval_results["rel+sugg"][key] += 1
        if rel_type == "part_of":
            eval_results["rel+part"][key] += 1
            
        if pred_rel in all_rels:
            key = "num_correct_label"
            eval_results["rel"][key] += 1
            if subj in corr_pred_ner_indices and obj in corr_pred_ner_indices:
                eval_results["rel+"][key] += 1
                if rel_type == "modify":
                    eval_results["rel+modify"][key] += 1
                if rel_type == "located_at":
                    eval_results["rel+locat"][key] += 1
                if rel_type == "suggestive_of":
                    eval_results["rel+sugg"][key] += 1
                if rel_type == "part_of":
                    eval_results["rel+part"][key] += 1
            
    
for eval_field, result_dict in eval_results.items():
    num_corr = result_dict["num_correct_label"]
    num_pred = result_dict["num_pred_label"]
    num_gold = result_dict["num_gold_label"]
    p = num_corr / num_pred if num_corr > 0 else 0.0
    r = num_corr / num_gold if num_corr > 0 else 0.0
    f1 = 2 * (p * r) / (p + r) if num_corr > 0 else 0.0
    print(f"[{eval_field}]: P: {p:.5f}, R: {r:.5f}, 【F1: {f1 * 100:.3f}】")
```
